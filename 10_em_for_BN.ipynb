```python
# Importing necessary libraries
import pandas as pd
from sklearn.datasets import load_iris
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import ExpectationMaximization, BayesianEstimator

# Loading the Iris dataset
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Adding a categorical target variable for Bayesian modeling
df['target'] = data.target

# Defining the Bayesian Network structure
model = BayesianNetwork([('sepal length (cm)', 'target'), 
                         ('sepal width (cm)', 'target'),
                         ('target', 'petal length (cm)'), 
                         ('target', 'petal width (cm)')])

# Applying the EM algorithm to estimate parameters
em = ExpectationMaximization(model, df)

# Estimating Conditional Probability Distributions (CPDs)
cpds = em.get_parameters()

# Displaying the learned parameters
for cpd in cpds:
    print(cpd)
```
